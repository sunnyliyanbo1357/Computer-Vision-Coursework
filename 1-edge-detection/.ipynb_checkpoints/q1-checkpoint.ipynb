##Task 1: Image filtering and enhancement

###Question 1

```python
%matplotlib inline
import numpy as np
import scipy
from scipy import signal
from scipy import misc
from matplotlib import pylab as plt
import matplotlib.cm as cm

x = np.array([[4, 1, 6, 1, 3],
            [3, 2, 7, 7, 2],
            [2, 5, 7, 3, 7],
            [1, 4, 7, 1, 3],
            [0, 1, 6, 4, 4]])
k = (1/9.0)*np.array([[1, 1, 1], 
                    [1, 1, 1], 
                    [1, 1, 1]])
signal.convolve2d(x, k, boundary='symm', mode='same')
```

###Question 2

```python
scipy.ndimage.filters.median_filter(x, 3)
```

It is seen that the in the central $3x3 for the mean filter is all close to 4 in terms of value. While for the mean filter, there is still some variation in the values as it changes from 4 to 7, while in the original it went from 2 to 7. Hence the median filter preserves the edges in the image better. Also, it is seen that all the numbers in the output are still whole numbers after convolving with the median filter.

###Question 3

```python
import math
sobx = np.array([[-1, 0, 1], 
                [-2, 0, 2], 
                [-1, 0, 1]])
soby = -1*np.transpose(sobx)

df_dx = signal.convolve2d(x, sobx, boundary='symm', mode='same')
df_dy = signal.convolve2d(x, soby, boundary='symm', mode='same')
df_dx_center = df_dx[2][2]
df_dy_center = df_dy[2][2]
gradient_magnitude_ctr = np.sqrt(np.square(df_dx_center)+np.square(df_dy_center))
gradient_direction_ctr = math.degrees(math.atan(df_dy_center/df_dx_center*1.0))

print "Gradient magnitude at center:", gradient_magnitude_ctr
print "Gradient direction at center (in degrees):", gradient_direction_ctr
```

###Question 5

```python
from scipy import ndimage
from scipy import misc

def display(img):
    plt.imshow(img, cmap = cm.Greys_r)
    plt.show()
    
def blur(img, g_sigma, n=2):
    print "Blur for sigma =", g_sigma
    return ndimage.filters.gaussian_filter(img, g_sigma)

def f_unsharp(img, g_sigma, n=2):
    g = ndimage.filters.gaussian_filter(img, g_sigma)
    print "Unsharp for sigma =", g_sigma
    unsharp_img = n*img - g
    return unsharp_img



img1 = misc.imread("./Images/Q1/cameraman.jpg")
img2 = misc.imread("./Images/Q1/house.jpg")
img3 = misc.imread("./Images/Q1/lena.jpg")

display(img1)
display(img2)
display(img3)
```

```python
display(f_unsharp(blur(img1, 0.75), 0.75))
display(f_unsharp(blur(img1, 2.5), 2.5))
```

```python
display(f_unsharp(blur(img2, 0.75), 0.75))
display(f_unsharp(blur(img2, 2.5), 2.5))
```

```python
display(f_unsharp(blur(img3, 0.75), 0.75))
display(f_unsharp(blur(img3, 2.5), 2.5))
```

We can see for sigma = 2.5, we get a highly sharpened image. This is because the gaussian filter acts as a low pass filter filtering out high frequencies. When you subtract the blurred image from the original, you only get the high frequency components (shown below). Adding the high frequency components to the original image emphasizes the edges strongly.

For sigma = 2.5, it is a more spread out gaussian than for sigma = 0.75. So, this results in the final image having thicker edges in the case of sigma = 2.5 than for sigma = 0.75

```python
display(f_unsharp(img3, 0.75, 1))
display(f_unsharp(img3, 2.5, 1))

display(f_unsharp(blur(img3, 0.75), 0.75, 1))
display(f_unsharp(blur(img3, 2.5), 2.5, 1))
```

```python

def quantize_kmeans(img, k):
    # Convert from 8 bits integer encoding to floats and divide by 255
    img = np.array(img, dtype=np.float64) / 255
    w, h = img.shape[0], img.shape[1]
    img_flat = np.reshape(img, (w*h, 3))
    img_flat_sample = shuffle(img_flat, random_state=0)[:5000]
    kmeans = KMeans(n_clusters=k, random_state=0).fit(img_flat_sample)
    labels = kmeans.predict(img_flat)
    quantized_rgbs = kmeans.cluster_centers_
    img_quantized = np.zeros(img.shape)
    for i in range(w):
        for j in range(h):
            img_quantized[i][j] = quantized_rgbs[labels[i*h+j]]
    return img_quantized

img = misc.imread("./Images/Q2/colorful2.jpg")
imgq = quantize_kmeans(img, 64)
display(img)
display(imgq)
```

###Question 2

```python
def quantize_lab(rgb, k, retlab = False):
    w, h = rgb.shape[0], rgb.shape[1]
    lab = color.rgb2lab(rgb)
    lab_flat = np.reshape(lab, (w*h, 3))
    lab_flat_sample = shuffle(lab_flat, random_state=0)[:5000]
    kmeans = KMeans(n_clusters=k, random_state=0).fit(lab_flat_sample)
    labels = kmeans.predict(lab_flat)
    quantized_labs = kmeans.cluster_centers_
    lab_quantized = np.array(lab, copy=True)
    for i in range(w):
        for j in range(h):
            lab_quantized[i][j][0] = quantized_labs[labels[i*h+j]][0]

    if retlab:
        return lab_quantized
    
    rgb_quantized = color.lab2rgb(lab_quantized)
    return rgb_quantized
```

```python
from skimage import io, color
rgb = io.imread("./Images/Q2/colorful2.jpg")
rgbq = quantize_lab(rgb, 16)
display(rgb)
display(rgbq)
```

```python
def plot_histogram(rgb, k):
    w, h = rgb.shape[0], rgb.shape[1]
    lab = color.rgb2lab(rgb)
    l = lab[:,:,0].reshape(w*h,)
    plt.hist(l,bins=k,range=(0,100))
    plt.show()
    
    labq = quantize_lab(rgb, k, True)
    lq = labq[:,:,0].reshape(w*h,)
    unique, counts = np.unique(lq, return_counts=True)
#     plt.plot(unique, counts, 'ro')
    plt.vlines(unique, 0, counts, color='k', linestyles='solid')
    plt.axis([0, 100, 0, np.max(counts)*1.1])
    plt.show()
#     plt.hist(lq,bins=100,range=(0,100))
#     plt.show()
```
